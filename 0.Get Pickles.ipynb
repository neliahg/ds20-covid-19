```python
from pathlib import Path
import pandas as pd

base_dir = Path.cwd()
if base_dir.name == "pickles":
    base_dir = base_dir.parent  

source_dir = base_dir / "source_files"
target_dir = base_dir / "pickles"
target_dir.mkdir(exist_ok=True)

for csv_file in source_dir.glob("*.csv"):
    df = pd.read_csv(csv_file)
    df.to_pickle(target_dir / (csv_file.stem + ".pkl"))
    print(f"Converted: {csv_file.name} -> {csv_file.stem}.pkl")

```
